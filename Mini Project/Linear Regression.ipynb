```python
dataset=data.drop(['Density', 
                   'Weight','Height','Neck','Chest','Abdomen','Hip','Thigh','Knee','Ankle','Biceps','Forearm','Wrist'],axis=1)
```

```python
x=dataset.iloc[:,0].values.reshape(-1,1)
```

### Data Visualization

```python
plot.scatter(x,y)
plot.xlabel('BodyFat')
plot.ylabel('Age')
plot.show
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```
